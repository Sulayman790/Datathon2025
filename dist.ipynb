```python
import tarfile, boto3
from pathlib import Path

bucket = "lawdemo-input"
key = "model.tar.gz"

src = Path("code")
dst = Path("dist/model.tar.gz")
dst.parent.mkdir(parents=True, exist_ok=True)

with tarfile.open(dst, "w:gz") as t:
    t.add(src, arcname="code")

boto3.client("s3").upload_file(str(dst), bucket, key)
print(f"s3://{bucket}/{key}")
```

```python
import boto3

s3 = boto3.client('s3')

def create_output_file(job_id: str, content: str):
    bucket = 'lawdemo-output'
    key = f'{job_id}/anything.jpg'
    s3.put_object(Bucket=bucket, Key=key, Body=content.encode('utf-8'), ContentType='text/plain')
    return f's3://{bucket}/{key}'

if __name__ == '__main__':
    job_id = 'LAW-20251102-230722-SXBE'
    uri = create_output_file(job_id, f'Job ID: {job_id}')
    print('Created:', uri)
```
